```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv("salary_data_cleaned.csv")
df.head()
```

```python
df.columns
```

```python
def tittle_simplifier(tittle):
    if "data scientist" in tittle.lower():
        return "data scientist"
    elif "data engineer" in tittle.lower():
        return "data engineer"
    elif "analyst" in tittle.lower():
        return "analyst"
    elif "machine learning" in tittle.lower():
        return "mle"
    elif "manager" in tittle.lower():
        return "manager"
    elif "director" in tittle.lower():
        return "director"
    else:
        return "na"


def seniority(tittle):
    if "sr" in tittle.lower() or "senior" in tittle.lower() or "lead" in tittle.lower() or "principal" in tittle.lower():
        return "senior"
    elif "jr" in tittle.lower() or "junior" in tittle.lower():
        return "junior"
    else:
        return "na"
    

```

```python
df["job_simfly"]=df["Job Title"].apply(tittle_simplifier)
df["seniority"]=df["Job Title"].apply(seniority)
```

```python
df["job state"]=df.job_state.apply(lambda x: x.strip() if x.strip().lower() != "los angeles" else "CA")
df["job state"].value_counts()
```
